```python
# Cargar librerías
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Cargar datos
np.random.seed(0)
horasdeestudio = np.random.rand(100, 1) * 10
calif = 2 * horasdeestudio + np.random.randn(100, 1) * 2

# Preprocesamiento de datos
train_data, test_data, train_labels, test_labels = train_test_split(horasdeestudio, calif, test_size=0.2, random_state=0)

# Selección y entrenamiento del modelo
modelo = LinearRegression()
modelo.fit(train_data, train_labels)

# Prueba de modelo
predic = modelo.predict(test_data)

# Graficar modelo de regresión lineal
plt.scatter(test_data, test_labels, color='blue')
plt.plot(test_data, predic, color='red')
plt.title('Modelo de Regresión Lineal')
plt.xlabel('Horas de estudio')
plt.ylabel('Calificaciones')
plt.show()

# Calcular el error cuadrático medio
mse = mean_squared_error(test_labels, predic)
print("Error cuadrático medio:", mse)
```
